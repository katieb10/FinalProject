```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import io
from bokeh.plotting import figure, show
from bokeh.layouts import row
from bokeh.io import output_notebook, show

```

```python
# import file
from google.colab import files 
uploaded = files.upload()
```

```python
# show data
df = pd.read_csv(io.BytesIO(uploaded['zoo2.csv']))
print(df)
```

```python
#clean data
df.drop_duplicates(inplace=True) #drop duplicates
#drop useless columns
df.drop(["toothed"], axis=1, inplace=True)
df.drop(["hair"], axis=1, inplace=True)
df.drop(["feathers"], axis=1, inplace=True)
df.drop(["eggs"], axis=1, inplace=True)
df.drop(["milk"], axis=1, inplace=True)
df.drop(["backbone"], axis=1, inplace=True)
df.drop(["breathes"], axis=1, inplace=True)
df.drop(["venomous"], axis=1, inplace=True)
df.drop(["fins"], axis=1, inplace=True)
df.drop(["legs"], axis=1, inplace=True)
df.drop(["tail"], axis=1, inplace=True)
df.drop(["domestic"], axis=1, inplace=True)
df.drop(["catsize"], axis=1, inplace=True)
df.drop(["class_type"], axis=1, inplace=True)
```

```python
# show clean data
# print(df)
x = ['airborne', 'aquatic']
y=[sum(df['airborne']), sum(df['aquatic'])]
plt.bar(x, y)
```

```python
output_notebook()
```

```python
#create bokeh plot
from bokeh.models import ColumnDataSource
source_Q1 = ColumnDataSource(df)
from bokeh.io import output_notebook, show
output_notebook()
from bokeh.plotting import figure 
p = figure(title = 'Predators in Aquatic and Airborne Environments')
p.circle("predator", "aquatic", source = source_Q1, color = 'green', size = 8, legend = dict(value = "aquatic"))
p.triangle("predator", "airborne", source = source_Q1, color = 'blue', size = 8, legend = dict(value = "airborne"))
p.legend.location = "top_left"
show(p)
```
